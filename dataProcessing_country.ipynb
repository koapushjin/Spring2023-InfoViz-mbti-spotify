```
df['I'] = df['INFP']+df['INFJ']+df['INTP']+df['INTJ']+df['ISFP']+df['ISFJ']+df['ISTP']+df['ISTJ']
df['E'] = df['ENFP']+df['ENFJ']+df['ENTP']+df['ENTJ']+df['ESFP']+df['ESFJ']+df['ESTP']+df['ESTJ']

df['N'] = df['INFP']+df['INFJ']+df['INTP']+df['INTJ']+df['ENFP']+df['ENFJ']+df['ENTP']+df['ENTJ']
df['S'] = df['ISFP']+df['ISFJ']+df['ISTP']+df['ISTJ']+df['ESFP']+df['ESFJ']+df['ESTP']+df['ESTJ']

df['F'] = df['INFP']+df['INFJ']+df['ISFP']+df['ISFJ']+df['ENFP']+df['ENFJ']+df['ESFP']+df['ESFJ']
df['T'] = df['INTP']+df['INTJ']+df['ISTP']+df['ISTJ']+df['ENTP']+df['ENTJ']+df['ESTP']+df['ESTJ']

df['P'] = df['INTP']+df['INFP']+df['ISTP']+df['ISFP']+df['ENTP']+df['ENFP']+df['ESTP']+df['ESFP']
df['J'] = df['INTJ']+df['INFJ']+df['ISTJ']+df['ISFJ']+df['ENTJ']+df['ENFJ']+df['ESTJ']+df['ESFJ']
```
